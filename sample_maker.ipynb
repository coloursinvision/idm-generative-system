```python
import numpy as np
import soundfile as sf

SAMPLE_RATE = 44100
```

```python
def normalize(x):
    return x / np.max(np.abs(x))
```

```python
def glitch_click(length_ms=10):
    length = int(SAMPLE_RATE * length_ms / 1000)
    noise = np.random.randn(length)
    envelope = np.exp(-np.linspace(0, 6, length))
    return normalize(noise * envelope)
```

```python
def noise_burst(length_ms=80, tone=0.3):
    length = int(SAMPLE_RATE * length_ms / 1000)
    noise = np.random.randn(length)
    
    t = np.linspace(0, 1, length)
    envelope = np.exp(-6 * t)
    
    low_noise = np.convolve(noise, np.ones(10)/10, mode="same")
    
    return normalize((1-tone)*noise + tone*low_noise) * envelope
```

```python
def fm_blip(freq=300, mod_freq=80, mod_index=2, length_ms=80):
    length = int(SAMPLE_RATE * length_ms / 1000)
    t = np.linspace(0, length_ms/1000, length)
    
    modulator = np.sin(2 * np.pi * mod_freq * t) * mod_index
    carrier = np.sin(2 * np.pi * freq * t + modulator)
    
    envelope = np.exp(-6 * t)
    return normalize(carrier * envelope)
```

```python
def save_sample(x, name):
    sf.write(name, x, SAMPLE_RATE)

save_sample(glitch_click(), "glitch_click.wav")
save_sample(noise_burst(), "noise_burst.wav")
save_sample(fm_blip(), "fm_blip.wav")
```

```python
for i in range(1, 9):
    save_sample(glitch_click(np.random.randint(5, 25)), f"A{i}_glitch.wav")
    save_sample(noise_burst(np.random.randint(40, 120)), f"B{i}_noise.wav")
    save_sample(fm_blip(
        freq=np.random.randint(150, 1200),
        mod_freq=np.random.randint(40, 300),
        mod_index=np.random.uniform(0.5, 6)
    ), f"C{i}_fm.wav")
```
